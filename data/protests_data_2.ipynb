```python
import pandas as pd
import numpy as np
import nltk
import nltk
nltk.download('punkt')
```

```python
protests = pd.read_csv('protests.csv', encoding = "latin")
```

```python
import re
from nltk.stem.snowball import SnowballStemmer
from nltk.tokenize import word_tokenize

stemmer = SnowballStemmer('english')

#size 

def find_number(some_string):
    indicators = ['people', 'protesters', 'protestors', 'students','residents']
    tokens = nltk.word_tokenize(re.sub(r'[^\w\s]', '', some_string.strip().lower()))
    for i in range(len(tokens)):
        if tokens[i] in indicators:
            if tokens[i - 1].isnumeric():
                return int(tokens[i - 1])
            if tokens[i - 2].isnumeric():
                return int(tokens[i - 2])
            if tokens[i - 2].isnumeric():
                return int(tokens[i - 3])
            if tokens[i] == "hundreds":
                return "+100"
            if tokens[i] == "dozens":
                return "+12"
    return "Unknown"


#date 

def find_date(some_string):
    indicators = ['may', 'june', 'july', "september"]
    tokens = nltk.word_tokenize(re.sub(r'[^\w\s]', '', some_string.strip().lower()))
    
    for i in range (len(tokens)):
        if tokens[i] in indicators:
            if tokens[i+1].isnumeric():
                return str(tokens[i]).capitalize()+" "+str(tokens[i+1])+", 2020"
    return "Unknown"        
    

## Violence

def violence(some_string):
    indicators = ['gun', 'violent', 'violence', "firearm", "tear gas","fired","shots","fire"]
    tokens = nltk.word_tokenize(re.sub(r'[^\w\s]', '', some_string.strip().lower()))
    for i in range (len(tokens)):
        if tokens[i] in indicators:
            return "Yes"
    return "Unknown" 


## Property Damage

def damage(some_string):
    indicators = ["damage", "damaged" , "burglaries" ,"burglary","smashed","smash"]
    tokens = nltk.word_tokenize(re.sub(r'[^\w\s]', '', some_string.strip().lower()))
    for i in range (len(tokens)):
        if tokens[i] in indicators:
            return "Yes"
    return "Unknown" 

# n-causalties  (injuries + fatalities)
examples = ["dozens of arrests", "four arrests", "one protester died",
            "a 21-year-old Eastpointe man in the vehicle was hit and died from his injuries"]

# Calling functions created 

protests['size'] = protests['description'].apply(lambda x: find_number(x))
protests['date'] = protests['description'].apply(lambda x: find_date(x))
protests['violence'] = protests['description'].apply(lambda x: violence(x))
protests['damage'] = protests['description'].apply(lambda x: damage(x))

protests
```
